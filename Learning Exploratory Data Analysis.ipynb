## Understanding Group By

Aggregates the variables with respect to the given parameters.

```python
df_final.columns
```

## Task: Aggregate -- Aggregaterating, Rating Color, Rating Text with respect to countries

```python
df_final.groupby(['Aggregate rating', 'Rating color','Rating text']).size() 

#Groups the three columns and returns the count in each according to the grouped variable. 
#Each row in the output is the unique combination of three variables and any repetition in these combinations results
#in an increase in the count in the right most column.

# Grouping concept is same as in sql
```

## Resetting the index and renaming a column name in groupby

```python
df_final.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index()
# Why reset?
# By resetting the index we now have a name of each column and a row starting from 0. 
# Thus we have an access path to each of the rows and columns of the dataframe.
```

## Renaming

```python
ratings_df=df_final.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
# Since the column was set to 0 using the reset index in last step we can easily perform the rename operation
ratings_df
```

# Hue Parameter

Hue parameter in seaborns library plots are used to add third dimension to the same 2-D graph.
This is done by colour encoding of the third dimension. Thus the third dimension should preferably be categorical and the categories should be less for clear distinction.

```python
sns.barplot(x="Aggregate rating",y="Rating Count", data=ratings_df, hue= "Rating color"); 
```

Observe that the rating color has been assigned colours automatically randomly and not as defined in the definition. This can be mapped using a parameter palette.

```python
sns.barplot(x="Aggregate rating",y="Rating Count", data=ratings_df, hue= "Rating color", palette=("white","red","orange","yellow","green", "green")); 
#Since 0.0 is set to white we cannot see it
```
